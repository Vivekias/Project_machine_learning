## Spilit the Data into x and y

```python
x=df.drop('target',axis=1)
y=df['target']
print(x.shape,y.shape)
```

## Preprocessing the Data using Standard Scaler:-

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_sc=sc.fit_transform(x)
print(x_sc)
print(y)
```

## Train and Test data split:-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
print(f'x_train={x_train.shape},x_test={x_test.shape}')
print(f'y_train={y_train.shape},y_test={y_test.shape}')
```

## Creating a function to compute classification report,confusion matrix,accuracy score:-

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_mat(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print(f'Accuracy Score={acc}')
    cm=confusion_matrix(y_test,y_pred)
    print(f'Confusion Matrix={cm}')
    print("Classification Report",classification_report(y_test,y_pred))
    
def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print('Training Score=',train_score)
    print('Testing Score=',test_score)
```

## Build The Models:-

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

### logistic Regression Model:-

```python
lr=LogisticRegression(max_iter=1000,solver='liblinear')
lr.fit(x_train,y_train)
```

```python
## Model score:-

mscore(lr)
```

```python
## Prediction on y values:-

y_pred_lr=lr.predict(x_test)
print(y_pred_lr)
```

```python
## Evaltion Metrics:-

eval_mat(y_test,y_pred_lr)
```

### K Neighbors Classifier Model:-

```python
knc=KNeighborsClassifier(n_neighbors=3)
knc.fit(x_train,y_train)
```

```python
## Model Score:-

mscore(knc)
```

```python
## Prediction on y values:-

y_pred_knc=knc.predict(x_test)
print(y_pred_knc)
```

```python
## Evalution Metrics:-

eval_mat(y_test,y_pred_knc)
```

### Decision Tree Classifier Model:-

```python
dt=DecisionTreeClassifier(criterion='entropy',max_depth=15,min_samples_split=25)
dt.fit(x_train,y_train)
```

```python
## Model Score:-

mscore(dt)
```

```python
## Prediction on y values:-

y_pred_dt=dt.predict(x_test)
print(y_pred_dt)
```

```python
## Evalution Metrics:-

eval_mat(y_test,y_pred_dt)
```

```python
fn=x_train.columns
cn=['Yes','No']  ## here 0= Yes and 1= No
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(dt,class_names=cn,feature_names=fn,filled=True)
plt.show()
```

### Ensemble as Bagging Classifier Model:-

```python
ebc=BaggingClassifier(base_estimator=dt,n_estimators=15,max_features=x.shape[1],random_state=25)
ebc.fit(x_train,y_train)
```

```python
## Model Score:-

mscore(ebc)
```

```python
## Prediction on y values:-

y_pred_ebc=ebc.predict(x_test)
print(y_pred_ebc)
```

```python
## Evalution Metrics:-

eval_mat(y_test,y_pred_ebc)
```

### Ensemble as Ada Boost Classifier Model:-

```python
ada=AdaBoostClassifier(base_estimator=dt,n_estimators=1)
ada.fit(x_train,y_train)
```

```python
## Model Score:-

mscore(ada)
```

```python
## Prediction on y values:-

y_pred_ada=ada.predict(x_test)
print(y_pred_ada)
```

```python
## Evalution Metrics:-

eval_mat(y_test,y_pred_ada)
```

## Checking best model:-

```python
check_best=pd.DataFrame({'Model Name':['Logistic Regression','KNeighbors Classifier','Decision Tree','Bagging Classifier','AdaBoost Classifier'],'Model Testing Score':[lr.score(x_test,y_test),knc.score(x_test,y_test),dt.score(x_test,y_test),ebc.score(x_test,y_test),ada.score(x_test,y_test),],'Model Traning Score':[lr.score(x_train,y_train),knc.score(x_train,y_train),dt.score(x_train,y_train),ebc.score(x_train,y_train),ada.score(x_train,y_train)]})
check_best
```
